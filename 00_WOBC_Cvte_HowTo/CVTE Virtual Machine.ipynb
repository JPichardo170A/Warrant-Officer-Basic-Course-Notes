# How to access your CVTE Virtual Machines at home


## 1. Download VMWare Horizon Client

1. Download and install the VMWare Horizon Client (you will likely need to restart your system).
2. Open the Omnissa Horizon Client.
3. Add Server: `https://vdi.ccoelearning.net`
4. Open Chrome (it should default to `home.ccoecaas.net/SAAS/auth/login`)
   - Domain: `ccoecaas.net`
5. Click **Next** and **Agree**.
6. Complete login with CAC credentials.
7. **Do NOT resize** this VM window if you plan to use multiple monitors using RDP.

At this point, you have two options for accessing your VM:


---

### Option A: Using the built-in remote console
_(Note: This method does **NOT** support multiple display screens.)_

1. Once logged in, select **Open for Aria Automation: CSTN – Production**.
2. Click on **Service Broker**.
3. On the left, click **Deployments**.
4. Click your Ops Station (listed with your name).
5. In the Topology window on the left, select the OS you want to use (in class, typically Ubuntu).
6. On the right, click **Actions** → **Connect to Remote Console**.
7. A new tab will open and launch your VM.

---

### Option B: Using Remote Desktop Connection (RDC)
_(Supports multiple display screens.)_

1. Follow steps 1–5 under “Using the built-in remote console” to note the VM’s IP address.
2. Once Windows VM is running, open **Remote Desktop Connection** from the Start menu.
3. Enter the previously recorded VM IP address.
4. Click **Show Options**, then go to the **Display** tab.
5. Check **Use all my monitors for the remote session**.
6. Click **Connect**.
   - If you see a blank screen, press **Enter** or click inside it (the VM may be asleep).
